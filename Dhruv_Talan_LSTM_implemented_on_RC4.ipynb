```python
mlp.columns=['hex','2nd_byte']
```

```python
def is_hex(s):
    try:
        int(s, 16)
        return True
    except ValueError:
        return False

for index, row in mlp.iterrows():
    if not is_hex(row['hex']):
        mlp.drop(index, inplace=True)
```

```python
mlp
```

```python
def hex_to_bin(hex_num):
    return bin(int(hex_num, 16))[2:].zfill(40)

mlp['Binary'] = mlp['hex'].apply(hex_to_bin)
```

```python
mlp
```

```python
mlp.drop('hex', axis=1, inplace=True)
```

```python
mlp
```

```python
def hex_to_bin(hex_num):
    return int(hex_num, 16)
mlp['2BYTE'] = mlp['2nd_byte'].apply(hex_to_bin)
```

```python
mlp.drop('2nd_byte', axis=1, inplace=True)
```

```python
mlp
```

```python
import numpy as np

def int_to_vector(integer):
    vector = np.zeros(256)
    vector[integer] = 1
    return vector
```

```python
mlp['ARR_REP'] = mlp['2BYTE'].apply(int_to_vector)
```

```python
mlp.drop('2BYTE', axis=1, inplace=True)
```

```python
mlp
```

```python
def convert_to_array(s):
    # convert your 40-bit string to an array of 40 elements each having strings elements as 0's and 1's
    x = np.array([float(i) for i in s])
    return x
mlp['Binary'] = mlp['Binary'].apply(convert_to_array)
```

```python
# convert the 40-bit strings to numpy arrays
X = np.array(mlp['Binary'].apply(list).tolist())

# # convert the 256-length vectors to numpy arrays
y = np.array(mlp['ARR_REP'].apply(list).tolist())

# # reshape the input data to match the LSTM input shape
X = X.reshape(X.shape[0], X.shape[1], 1)
```

```python
X.shape
```

```python
y.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
# train the model on the input data
model = Sequential()
model.add(LSTM(256, input_shape=(40, 1)))
model.add(Dense(256, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
model.fit(X, y, epochs=10, batch_size=32)
```
